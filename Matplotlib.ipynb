```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
q=[1,3,5,7,9]
q2=[9,1,4,5,6]
plt.xlabel('day')
plt.ylabel('weather')
plt.plot(q,q2,color='g',marker='H',linewidth=2, markersize=12,alpha=0.5)
```

```python
l1=[1,2,3,4,5]
l2=[10,20,70,8,10]
l3=[3,6,9,12,15]
l4=[4,8,12,16,20]
```

```python
plt.plot(l1,l2,'g',label='min')
plt.plot(l3,l4,'r',label='max')
plt.plot(l1,l3,'b',label='avg')
plt.xlabel('rate')
plt.ylabel('price')
plt.title('stock')
plt.legend(loc="best",shadow=True,fontsize='small')
plt.grid()
```

BAR CHART


```python
cop=['google','microsoft','amazon']
nift=[8,1,5]
profit=[6,3,2]
```

```python
t=np.arange(len(cop))
plt.bar(t,nift,label='top')
plt.bar(t,profit,label='profit')
plt.xticks(t,cop)
plt.xlabel('company')
plt.ylabel('nift')
plt.title('stock')
plt.legend()
```

```python
t=np.arange(len(cop))
plt.bar(t-0.2,nift,width=0.5,label='top')
plt.bar(t+0.2,profit,width=0.5,label='profit')
plt.xticks(t,cop)
plt.xlabel('company')
plt.ylabel('nift')
plt.title('stock')
plt.legend()
```

```python
plt.bar(t-0.2,nift,width=0.5,label='top')
plt.bar(t+0.2,profit,width=0.5,label='profit')
plt.yticks(t,cop)
plt.xlabel('company')
plt.ylabel('nift')
plt.title('stock')
plt.legend()
```

#pie chart


```python
exp=[1400,200,1000]
t=['washing','grocery','travel']
```

```python
plt.axis('equal')
plt.pie(exp,labels=t,radius=1.5)
```

```python
plt.axis('equal')
plt.pie(exp,labels=t,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0,0,1])
```
